# 3D SCATTER PLOT

Bazı durumlarda 2D data'yı anlamak için yeterli olmuyor. Bu durumda bir boyut daha eklemek data'nın anlaşılabilirliği artırabilir. 3 boyutlu x,y,z ekseninde bir plot var. Rengi eklersek 4 boyuta çıkar ve birde scatter'ların point'lerinin size'ını eklersek 5 boyuta çıkar. Ama aşağıdaki uygulamamızda size'ı eklemicez.

```python
from plotly.offline import iplot

dataframe = timesData[timesData.year == 2015]

trace1 = go.Scatter3d(
    x=dataframe.world_rank,
    y=dataframe.research,
    z=dataframe.citations,
    mode='markers',
    marker=dict(
        size=10,
        color='rgb(255,0,0)',
    )
)

data = [trace1]
layout = go.Layout(
    margin=dict(  #kenarlardan bırakılan boşluktur.
        l=0,
        r=0,
        b=0,
        t=0  
    )
    
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

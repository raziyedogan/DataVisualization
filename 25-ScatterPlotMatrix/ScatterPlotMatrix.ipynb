# SCATTER PLOT MATRIX

Scatter plot matrix, iki tane feature arasındaki ilişkiyi görmek için kullanılır.

```python
from plotly.offline import iplot 

# import figure factory
import plotly.figure_factory as ff
# prepare data
dataframe = timesData[timesData.year == 2015]
data2015 = dataframe.loc[:,["research","international", "total_score"]]
data2015["index"] = np.arange(1,len(data2015)+1)
# scatter matrix
fig = ff.create_scatterplotmatrix(data2015, diag='box', index='index',colormap='Portland',
                                  colormap_type='cat',
                                  height=700, width=700)
iplot(fig)
```

"research","international", "total_score" feature'leri, karşılaştırmak istediğimiz feature'lerdir. Dolayısıyla 
data2015 = dataframe.loc[:,["research","international", "total_score"]] satırı ile bunları bir data frame içerisine atıyoruz.

data2015["index"] = np.arange(1,len(data2015)+1) satırında data frame'in uzunluğunda bir liste hazırlayıp, kaç adet değer var ise 1'den uzunluk değerine kadar olan sayılar index sütununun her bir satırına sırasıyla atanır.

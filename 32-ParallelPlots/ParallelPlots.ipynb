```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import chart_studio.plotly as py
import plotly.graph_objs as go
import matplotlib_venn as venn
from math import pi
from pandas.plotting import parallel_coordinates
```

```python
data = pd.read_csv('Iris.csv')
```

```python
data.head()
```

Yukarıda gördüğünüzüzere 3 farklı tür vardır. 

```python
data = data.drop(['Id'],axis=1)
```

Id sütununu kullanmayacağımız için data'dan çıkardım. Ve veri seti şu hale geldi:

```python
data.head()
```

Yukarıda gördüğünüz üzere 4 tane feature vardır. Species sütununda kaç farklı tür olduğuna bakalım:

```python
data.Species.unique()
```

'Iris-setosa', 'Iris-versicolor', 'Iris-virginica' isimlerinde 3 tür bulunmaktadır.

```python
# Make the plot
plt.figure(figsize=(15,10))
parallel_coordinates(data, 'Species', colormap=plt.get_cmap("Set1"))
plt.title("Iris data class visualization according to features (setosa, versicolor, virginica)")
plt.xlabel("Features of data set")
plt.ylabel("cm")
plt.show()
```

Yukarıdaki kod bloğunu açıklayalım.

plt.figure(figsize=(15,10)) satırı ile x ekseni uzunluğu 15 ve y ekseni uzunluğu 10 olan figür oluşturdum.

parallel_coordinates(data, 'Species', colormap=plt.get_cmap("Set1")) satırında:

    1. parametre ile data değişkeninde tutulan veri setini kullanacağımızı belirttim.
    
    2. parametre ile veri setindeki Species sütununu belirttim.
    
    3. parametre ile Set1 isimli renk paletini kullanacağımızı belirttim. Bu renk paleti gri, kırmızı, turuncu renklerini içerir.
    
Böylece feature'lerin türler arasındaki değişimini görüntüledik.

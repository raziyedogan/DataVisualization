```python
timesData = pd.read_csv("timesData.csv")
```

# WORD CLOUD

Word cloud, plotly kütüphanesine ait değildir. Word cloud, görselleştirme açısından güzel bir tool olduğu için bunada değineceğiz.

2011 yılındaki üniversite sıralamasına göre ülkeleri baz alalım ve en çok hangi ülkenin kullanıldığını word cloud ile görselleştirelim.

```python
x2011 = timesData.country[timesData.year == 2011]
plt.subplots(figsize=(8,8))
wordcloud = WordCloud(
                          background_color='white',
                          width=512,
                          height=384
                         ).generate(" ".join(x2011))
plt.imshow(wordcloud)
plt.axis('off')

plt.show()
```

Yukarıdaki kod bloğu içerisinde bulunan .generate(" ".join(x2011)) kısmının anlamı şudur; Kullanılan kelimeler ayrılarak en çok yazılı olanların büyük bir şekilde plot ettirilmesidir.

plt.imshow(wordcloud) satırı ile bir resmin plot ettirilmesi sağlanır.

plt.axis('off') satırı ile x ve y eksenlerinin görünmesi engellenmiştir.
